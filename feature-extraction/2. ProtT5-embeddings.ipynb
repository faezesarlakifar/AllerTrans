```python
# @markdown Create the dataframe helper function
def make_df(embedding_path, label):
    f = h5py.File(embedding_path, 'r')
    keys = f.keys()
    df = pd.DataFrame()
    for key in keys:
        df = df.append([f[key]], ignore_index=True)
    labels = [label] * len(keys)
    df['Label'] = labels
    df['id'] = keys

    return df
```

## Save the train positive data as a .csv file

```python
h5_file_name = "positive_protT5_emb.h5"
embedding_path = output_path+"train_positive/"+h5_file_name
df_train_positive = make_df(h5_file_name, 1)
df_train_positive.to_csv('train_positive.csv')
df_train_positive.head()
```

```python
h5_file_name = "negative_protT5_emb.h5"
embedding_path = output_path+"train_negative/"+h5_file_name
df_train_negative = make_df(h5_file_name, 1)
df_train_negative.to_csv('train_negative.csv')
df_train_negative.head()
```

```python
h5_file_name = "protein_embeddings.h5"
embedding_path = output_path+"train_positive/"+h5_file_name
df_train_positive = make_df(embedding_path, 1)
df_train_positive.to_csv('train_positive.csv')
df_train_positive.head()
```
